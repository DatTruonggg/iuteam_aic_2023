```python
def calculate_cosine_distance(a, b):
    cosine_distance = float(spatial.distance.cosine(a, b))
    return cosine_distance

def calculate_cosine_similarity(a, b):
    cosine_similarity = 1 - calculate_cosine_distance(a, b)
    return cosine_similarity
```

```python
map_key_path ="D:\\DatTruong\\All\\2025\\HCM_AI\\Data\\Map_Chung\\"
base_path_to_plot = 'D:\\DatTruong\\All\\2025\\HCM_AI\\Data\\'


def get_path_frame(dict1,value):
    return dict1[str(value)][1]
def get_frame_info(dict1,value):
    key = dict1[str(value)][0]
    temp = key.split('-')
    df = pd.read_csv(map_key_path + temp[0]+'.csv')
    temp1 = [temp[0], df['frame_idx'][int((temp[-1]))]]
    return temp1

def plot_k_images(image_paths, K, info_imgs):

    grid_size = int((math.sqrt(K)))
    fig = plt.figure(figsize=(20,10))

  # Iterate over the image paths and plot the images
    for i in range(10, K + 1, grid_size * grid_size):
        for id, image_path in enumerate(image_paths[i - 10:i - 10 + grid_size * grid_size]):
            axes = fig.add_subplot(grid_size, grid_size, id + 1)
            axes.set_title(info_imgs[i - 10 + id])
            try:
                plt.imshow(Image.open(image_path))
            except Exception:
                pass



  # Tight layout and show the plot
    fig.tight_layout()
    plt.show()
```

```python
#Encode text descriptions -- CÁI NÀY LÀ QUERY INPUT --
#T DÙNG CÁI API TEXT 2 IMG CỦA GG ĐỂ LẤY CÂU QUERY NÀY CHO CÁI ẢNH THỬ 6 - 0007.jpg
def train(text_emb,k,dict_frame):

  #Compute cosine similarities
    id_list = []
    score_list = []
    cos_scores = util.semantic_search(torch.tensor(text_emb), torch.tensor(features,dtype=torch.float32),top_k = k)
    list_of_dictionaries = cos_scores[0]
    for item in list_of_dictionaries:
        id_list.append(item['corpus_id'])
        score_list.append(item['score'])

    path_imgs = []
    info_imgs = []
    print(id_list)
    print(list_of_dictionaries)
    for i in id_list:
        path_imgs.append(get_path_frame(dict_frame,i))
        info_imgs.append(get_frame_info(dict_frame,i))
    return path_imgs, info_imgs
```
